```python
import pandas as pd
```

```python
# Load the datasets
terrorism_df = pd.read_csv('/content/Global Terrorism Dataset 1970 - 2017.csv', encoding='latin1')
sp500_df = pd.read_csv('/content/S&P 500 Historical Data 1970-2024.csv')
```

```python
# head db
pd.set_option("display.max_columns", None)
display(terrorism_df.head())
display(sp500_df.head())
```

```python
# Calculate the percentage of NA values for each column
na_percentages = terrorism_df.isna().sum() / len(terrorism_df) * 100

# Identify columns with more than 50% NA values
columns_to_drop = na_percentages[na_percentages > 40].index

# Drop the identified columns
terrorism_df = terrorism_df.drop(columns=columns_to_drop)

# Display the updated DataFrame
display(terrorism_df.head())
print("Terrorism Dataset dimensions:", terrorism_df.shape)
print("S&P 500 Dataset dimensions:", sp500_df.shape)
```

```python
# Drop rows with any NA values in the terrorism dataset
terrorism_df.dropna(inplace=True)

# Display the updated DataFrame
display(terrorism_df.head())
print("Terrorism Dataset dimensions:", terrorism_df.shape)
```

```python
from ipyleaflet import Map, Marker, Popup
from ipywidgets import IntSlider, VBox, HTML
import pandas as pd

# Create a map centered on the US
m = Map(center=(40.7, -73.85), zoom=2)

# Create a slider for selecting the year
slider = IntSlider(description='Year', min=terrorism_df['iyear'].min(), max=terrorism_df['iyear'].max(), step=1)

# Function to update the map based on the selected year
def update_map(change):
    year = change['new']
    filtered_data = terrorism_df[terrorism_df['iyear'] == year]
    sample_size = min(len(filtered_data), 100)
    sample_data = filtered_data.sample(n=sample_size, random_state=42)

    # Clear existing markers
    m.layers = [m.layers[0]]  # Keep the base map layer

    # Add new markers
    for _, row in sample_data.iterrows():
        marker = Marker(location=(row['latitude'], row['longitude']))
        popup_text = (
            f"Year: {row['iyear']}<br>"
            f"Target: {row['target1']}<br>"
            f"Gang Name: {row['gname']}<br>"
            f"Weapon: {row['weaptype1_txt']}"
        )
        popup = Popup(child=HTML(value=popup_text), max_width=250)
        marker.popup = popup
        m.add_layer(marker)

# Link the slider to the map update function
slider.observe(update_map, names='value')

# Display the slider and the map
update_map({'new': slider.value})  # Initialize with the first year
VBox([slider, m])
```

```python
import plotly.express as px
import pandas as pd

sp500_df['Price'] = pd.to_numeric(sp500_df['Price'], errors='coerce')
sp500_df['High'] = pd.to_numeric(sp500_df['High'], errors='coerce')


sp500_df['Change %'] = pd.to_numeric(sp500_df['Change %'].astype(str).str.rstrip('%'), errors='coerce') / 100


sp500_df = sp500_df.dropna(subset=['High', 'Price', 'Change %'])


sp500_df['Date'] = pd.to_datetime(sp500_df['Date'])


fig = px.scatter(
    sp500_df,
    x='Change %',
    y='Date',
    size='Price',
    color='Price',
    animation_frame='Date',
    title='Interactive S&P 500 Scatter Plot'
)
fig.show()
```
